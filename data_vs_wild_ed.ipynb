```python
def lookup_dict(col_name):
    filter_=mushroom_prop_df.mush_feature==col_name
    lut=mushroom_prop_df[filter_]
    lud={}
    for i in range(len(lut)):
        lud[lut.feature_code.iloc[i]] = lut.feature_desc.iloc[i]
    return lud
```
